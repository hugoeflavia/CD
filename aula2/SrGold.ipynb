Seguem abaixo os gráficos de frequência absoluta e relativa para cada variável:

```python
import pandas as pd
import os
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')

meu_diretorio = "C:/Users/Hugo/Desktop/Ciência dos Dados"
os.chdir(meu_diretorio)
dados = pd.read_excel("PNAD.xlsx")

#tde = pd.crosstab(dados.V0404, dados.V6002) #coluna.value_counts()
#tde.plot(kind='bar', color =('blue','red'))
```

# Variável Sexo:

```python

coluna = dados.Sexo
coluna = coluna.astype('category')
coluna.cat.categories = ("Masculino","Feminino")
fabsoluta = coluna.value_counts()
frelativa = coluna.value_counts(True)*100
fabsoluta.plot(kind='bar', color=('red', 'blue'), legend=False)


```

```python
frelativa.plot(kind='pie',colors=('red','blue'), autopct="%0.2f",legend=False)
```

```python
coluna = dados.Cor
coluna = coluna.astype('category')
coluna.cat.categories = ("Branca","Preta","Amarela","Parda","Indigena","Sem declaração")
fabsoluta = coluna.value_counts()
frelativa = coluna.value_counts(True)*100
fabsoluta.plot(kind='bar', color=('red', 'blue','green','yellow','black'), legend=False)
```

```python
frelativa.plot(kind='pie',colors=('red', 'blue','green','yellow','black'), autopct="%0.2f",legend=False)
```
