# Análise de Regressão Múltipla

```python
from mpl_toolkits.mplot3d import Axes3D
import statsmodels.api as sm  #Usa OLS como no R
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm



from mpl_toolkits.mplot3d import Axes3D

X = dados[["idh","colesterol"]]
y = np.log(dados.eletricidade)

## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
est = sm.OLS(y, X).fit()

## Create the 3d plot -- skip reading this
# TV/Radio grid for 3d plot
xx1, xx2 = np.meshgrid(np.linspace(X.idh.min(), X.idh.max(), 100), 
                       np.linspace(X.colesterol.min(), X.colesterol.max(), 100))
# plot the hyperplane by evaluating the parameters on the grid
Z = est.params[0] + est.params[1] * xx1 + est.params[2] * xx2

# create matplotlib 3d axes
fig = plt.figure(figsize=(12, 8))
ax = Axes3D(fig, azim=-115, elev=15)

# plot hyperplane
surf = ax.plot_surface(xx1, xx2, Z, cmap=plt.cm.RdBu_r, alpha=0.6, linewidth=0)

# plot data points - points over the HP are white, points below are black
resid = y - est.predict(X)
ax.scatter(X[resid >= 0].idh, X[resid >= 0].colesterol, y[resid >= 0], color='black', alpha=1.0, facecolor='white')
ax.scatter(X[resid < 0].idh, X[resid < 0].colesterol, y[resid < 0], color='black', alpha=1.0)

# set axis labels
ax.set_xlabel('IDH')
ax.set_ylabel('Colesterol')
ax.set_zlabel('Energia consumida per capita')
```

```python
y=np.log(dados.eletricidade)
X=dados[["idh","colesterol"]]
X=sm.add_constant(X)  
X.head()

model=sm.OLS(y, X)
fit=model.fit()
fit.summary()
```

Variável resposta = Energia consumida per capita (ECPC)
Variáveis Explicativas = IDH e índice de colesterol no sangue em homens (ICS)
ECPC = 1.6350 + 9.0549*IDH + -0.1032*colesterol

Para verificar se as variáveis são relevantes ao modelo faremos um teste de hipótese, em que nossa hipótese nula e nossa hipótese alternativa serão respectivamente:

                                              H_0:  β ̂=0

                                              H_1: β ̂≠0

Analisando a tabela acima observamos que temos para o idh e o colesterol temos os respectivos valores-p: 0% e 43,5%. O que isso nos indica?
Que segundo o valor-p do idh a hipótese nula é completamente rejeitada, logo a variável é relevante ao modelo. Entretanto, quando observamos o valor-p do colesterol percebemos um valor elevado 43,5%, em teoria deveríamos aceitar Ho e assumir que a variável não é relevante ao modelo, porém, segundo nossas análises gráficas pode-se observar que isso não é verdade, nossa teoria é que esse valor-p encontrado é pelo fato da relação entre colesterol e energia consumida estar mais ligada a uma curva do que à uma reta, mas a relação não deixa de existir.
